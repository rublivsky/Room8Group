```python
import re
import pandas as pd
import numpy as np

path = "Mobiles Dataset (2025).csv"
```

```python
df = pd.read_csv(path, encoding="latin1")
```

```python
df.head()
```

```python
df.info()
```

Transform obj to int and remove different symbols

```python
def extract_number(value):
    if isinstance(value, str):
        value = re.sub(r'[^0-9.]', '', value.replace(',', ''))  # Remove non-numeric characters
        value = re.sub(r'\.(?=.*\.)', '', value)  # Keep only the first decimal point
        return float(value) if value else None
    return value

def extract_mp(value):
    if isinstance(value, str):
        mp_values = re.findall(r'\d+', value)  # Extract all numbers
        return float(mp_values[0]) if mp_values else None  # Take the first MP value
    return value

# Apply to both camera columns
df["Front Camera"] = df["Front Camera"].apply(extract_mp)
df["Back Camera"] = df["Back Camera"].apply(extract_mp)

# Clean numeric columns
df["RAM"] = df["RAM"].apply(extract_number)  # RAM in GB
df["Mobile Weight"] = df["Mobile Weight"].apply(extract_number)  # Weight in grams
df["Battery Capacity"] = df["Battery Capacity"].apply(extract_number)  # Battery in mAh
df["Screen Size"] = df["Screen Size"].apply(extract_number)  # Display size in inches

# Clean price columns
price_columns = ["Launched Price (Pakistan)", "Launched Price (India)", "Launched Price (China)",
                 "Launched Price (USA)", "Launched Price (Dubai)"]

for col in price_columns:
    df[col] = df[col].apply(extract_number)

# Save cleaned dataset
df.to_csv("Cleaned_Mobiles_Dataset.csv", index=False)
```
