```python
import re
import pandas as pd

# Load the two CSV files into separate DataFrames
games_df = pd.read_csv('/home/redmoon/Documents/Dev/Room8Group/data/VGInsights.csv')
platforms_df = pd.read_csv('/home/redmoon/Documents/Dev/Room8Group/data/GamePressure.csv')
```

```python
#Clean the  name column in both dataframes
games_df['name'] = games_df['name'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
platforms_df['name'] = platforms_df['name'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
```

```python
# Merge the two DataFrames on the key 'name'

# Добавляем новый столбец с названиями в нижнем регистре
games_df['name_lower'] = games_df['name'].str.lower()
platforms_df['name_lower'] = platforms_df['name'].str.lower()

# Выполняем объединение по новому столбцу
merged_df = pd.merge(games_df, platforms_df, on=['name_lower'], how='left')

# Удаляем временный столбец
merged_df = merged_df.drop(columns=['name_lower'])

merged_df.head()
```
