```python
import pandas as pd
from datetime import datetime, timedelta

df = pd.read_csv('../games_data.csv')
df.head()
```

```python
# Define the start date
start_date = datetime.now().strftime('%Y-%m-%d')

# Convert start_date to a datetime object
to_date = datetime.strptime(start_date, '%Y-%m-%d')

# Calculate the end date (2 years after the start date)
from_date = to_date - timedelta(days=2*365)
filtered_df = df[(pd.to_datetime(df['released']) >= from_date) & (pd.to_datetime(df['released']) <= to_date)]
filtered_df
```

```python
filtered_df.to_csv('filtered_games_data.csv', index=False, encoding='utf-8')
```
