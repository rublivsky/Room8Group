```python
import pandas as pd

# Загрузим датасет
df = pd.read_csv('/home/redmoon/Documents/Dev/Room8Group/data/PreFinal.csv')
dg_gm = pd.read_csv('/home/redmoon/Documents/Dev/Room8Group/data/GamePressureUpcoming.csv')
```

```python
df_platforms = df.assign(Platform=df['platforms'].str.split(', ')).explode('Platform')
df_platforms_gm = dg_gm.assign(Platform=dg_gm['platforms'].str.split(', ')).explode('Platform')
# Теперь каждая игра будет дублироваться по числу уникальных платформ
# df_platforms
df_platforms_gm
```

```python
df_platforms
```

```python
# Преобразуем столбец release_date в df_platforms_gm в формат datetime
df_platforms_gm['release_date'] = pd.to_datetime(df_platforms_gm['release_date'], errors='coerce')

# Создаем новый DataFrame с нужными столбцами из df_platforms_gm
df_platforms_gm_renamed = df_platforms_gm.rename(columns={
    'name': 'name',
    'release_date': 'released',
    'Platform': 'Platform',
    'genre': 'genre'
})[['name', 'released', 'Platform', 'genre']]

# Добавляем строки из df_platforms_gm_renamed в df_platforms
df_platforms = pd.concat([df_platforms, df_platforms_gm_renamed], ignore_index=True)
```

```python
df_platforms = df_platforms.drop(columns=['slug', 'vgi_id'])
```

```python
df_platforms
```

```python
df_platforms['released'] = pd.to_datetime(df_platforms['released'], errors='coerce').dt.strftime('%Y-%m-%d')
df_platforms = df_platforms[df_platforms['released'].str[:4].astype(float) <= 2025]
df_platforms = df_platforms[df_platforms['released'].str[:4].astype(float) >= 2024]
df_platforms['released'] = df_platforms['released'].replace('2025-01-01', '2025-12-31')
df_platforms['released'] = pd.to_datetime(df_platforms['released'], errors='coerce')
```

```python
# Решение "Оружие": Фильтруем игры с жанром Action
df_weapon = df_platforms[df_platforms['genre'] == 'Action']

# Решение "Портирование": Фильтруем игры, которые доступны только на PC или только на консолях
df_porting_pc = df_platforms[df_platforms['platforms'] == 'PC']
df_porting_consoles = df_platforms[df_platforms['platforms'].isin(['PlayStation', 'Xbox', 'Mobile', 'Nintendo'])]

# Решение "LiveOps": Создаем датафрейм для LiveOps (например, игры с большим количеством отзывов или активных пользователей)
df_liveops = df_platforms[df_platforms['reviews'] > 100000]
```

```python
# Добавляем тег "weapon" для игр из df_weapon
df_weapon['tag'] = 'weapon'

# Добавляем тег "porting" для игр из df_porting_pc и df_porting_consoles
df_porting_pc['tag'] = 'porting'
df_porting_consoles['tag'] = 'porting'

# Добавляем тег "liveops" для игр из df_liveops
df_liveops['tag'] = 'liveops'

# Объединяем все датафреймы
df_combined = pd.concat([df_weapon, df_porting_pc, df_porting_consoles, df_liveops], ignore_index=True)

df_combined.to_csv('combined_data.csv', index=False, encoding='utf-8')
```
